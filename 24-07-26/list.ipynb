## 1.2. `append()`

Adds a single item to the **end** of the list, in place.

```python
employees = ["John", "Alice"]
new_employee = input("Enter Employee Name: ")
new_employee =  new_employee.capitalize()
employees.append(new_employee)
print(employees)
```

```python
products = ["Laptop", "Mouse"]
products.append("Keyboard")
print(products)
```

```python
customers = []
customers.append("Ramesh")
customers.append("suresh")

customers = [name.capitalize() for name in customers]

print(customers)
```

```python
sales = []
sales.append(12000)
sales.append(15000)
print(sales)
```

## 1.4. `extend()`

Appends **all items** from another iterable onto the end of the list (unlike `append()`, which would add the whole iterable as one item).

```python
team_1 = ["John","Alice"]
team_2 = ["David","Rahul"]
team_1.extend(team_2)
print(team_1)
```

```python
electronics =["Laptop","TV"]
accessories = ["Mouse","Keyboard"]
electronics.extend(accessories)
print(electronics)
```

```python
classA = ["Ram", "Sam"]
classB = ["Tom", "Jerry"]

classA.extend(classB)

print(classA)
```

```python
hubli = ["Store1", "Store2"]
bangalore = ["Store3"]

hubli.extend(bangalore)

print(hubli)
```

## 1.5. `remove()`

Removes the **first matching value** from the list. Raises `ValueError` if the value isn't found.

```python
employees = ["John", "Alice", "Rahul"]

employees.remove("Alice")

print(employees)
```

```python
products = ["Laptop", "Mouse", "Keyboard"]

products.remove("Mouse")

print(products)
```

```python
orders = ["Order1", "Order2", "Order3"]

orders.remove("Order2")

print(orders)
```

```python
students = ["Ram","Sam","Tom"]
students.remove("Sam")
print(students)
```

## 1.6. `pop()`

Removes and **returns** the last item (or the item at a given index) — the only removal method that gives the value back.

```python
queue = ["A","B","C"]
served = queue.pop()
print(served)
print(queue)
```

```python
products = ["Laptop","TV","Mouse"]
products.pop()
print(products)
```

```python
tasks = ["Login", "Upload", "Logout"]

tasks.pop()

print(tasks)
```

```python
attendance = ["John", "Alice", "Rahul"]

attendance.pop()

print(attendance)
```

## 1.8. `sort()`

Sorts the list **in place**, ascending by default. Use `sort(reverse=True)` for descending order.

```python
employees = ["Rahul", "Alice", "John"]

employees.sort()

print(employees)
```

```python
prices = [250, 1000, 150, 600]

prices.sort()

print(prices)
```

```python
marks = [78, 45, 91, 66]

marks.sort()

print(marks)
```

```python
cities = ["Hubli", "Delhi", "Bangalore"]

cities.sort()

print(cities)
```

## 1.9. `reverse()`

Reverses the order of items **in place** (not a sort — just flips the current order).

```python
orders = ["Order1", "Order2", "Order3"]

orders.reverse()

print(orders)
```

```python
employees = ["John", "Alice", "David"]

employees.reverse()

print(employees)
```

```python
numbers = [10, 20, 30, 40]

numbers.reverse()

print(numbers)
```

```python
cities = ["Hubli", "Delhi", "Mumbai"]

cities.reverse()

print(cities)
```

## 1.12. `copy()`

Returns a **shallow copy** of the list. Without it, `backup = original` just points both names at the same list — mutating one mutates the other, as the last example below shows.

```python
employees = ["John", "Alice"]

backup = employees.copy()

print(backup)
```

```python
cart = ["Laptop", "Mouse"]

cart_backup = cart.copy()

print(cart_backup)
```

```python
attendance = ["Present", "Absent"]

attendance_backup = attendance.copy()

print(attendance_backup)
```

```python
products = ['tv','laptop']
#backup = products.copy()
backup = products
print(backup)
```

## 1.13. List Slicing

Extracts a sub-list using `list[start:stop]`. Negative indices count from the end, so `list[-2:]` grabs the last two items.

```python
employees = ["John", "Alice", "Rahul", "David", "Rohan"]

print(employees[:3])
```

```python
products = ["TV", "Laptop", "Mouse", "Keyboard"]

print(products[-2:])
```

```python
marks = [91, 88, 85, 80, 78, 75]

print(marks[:5])
```

```python
orders = ["O1", "O2", "O3", "O4", "O5"]

print(orders[-3:])
```

## 2. Mini Project — Student Management System

A menu-driven CLI that combines `append`, `index`, `remove`, and `len` to add, view, search, update, remove, and count students.

```python
students = []

while True:
    print("\n===== Student Management System =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Remove Student")
    print("6. Count Students")
    print("7. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter Student Name: ").title()

        if name not in students:
            students.append(name)
            print(name, "added successfully.")
        else:
            print("Student already exists.")

    elif choice == "2":
        if students:
            print("\nStudent List")
            for i, student in enumerate(students, start=1):
                print(i, student)
        else:
            print("No students available.")

    elif choice == "3":
        name = input("Enter Student Name to Search: ").title()

        if name in students:
            position = students.index(name) + 1
            print(name, "found at position", position)
        else:
            print("Student not found.")

    elif choice == "4":
        old_name = input("Enter Student Name to Update: ").title()

        if old_name in students:
            new_name = input("Enter New Student Name: ").title()
            index = students.index(old_name)
            students[index] = new_name
            print("Student details updated.")
        else:
            print("Student not found.")

    elif choice == "5":
        name = input("Enter Student Name to Remove: ").title()

        if name in students:
            students.remove(name)
            print(name, "removed successfully.")
        else:
            print("Student not found.")

    elif choice == "6":
        print("Total Students:", len(students))

    elif choice == "7":
        print("Thank you for using Student Management System.")
        break

    else:
        print("Invalid Choice.")
```
