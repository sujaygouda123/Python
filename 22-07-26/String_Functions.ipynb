```python
## 8. casefold() — Aggressive lowercase conversion for case-insensitive comparison
```

**Example 1**

```python
text = "Straße"

print(text.lower())
print(text.casefold())
```

**Example 2**

```python
stored_username = "sujay"

entered_username = input("Enter Username: ").casefold()

if entered_username == stored_username.casefold():
    print("Login Successful")
else:
    print("Invalid Username")
```

**Example 3**

```python
email1 = input("Enter the Email_1:")
email2 = input("Enter the Email_2:")

if email1.casefold() == email2.casefold():
    print("Email Matched")
else:
    print("Email Miss-Matched")
```

**Example 4**

```python
products = ["Laptop", "Mouse", "Keyboard"]

search = input("Enter Product Name: ")

for product in products:
    if product.casefold() == search.casefold():
        print("Product Found")
        break
else:
    print("Product Not Found")
```

```python
## 10. count() — Count occurrences of a substring
```

```python
The count() method returns the count or the number of times a particular value appears in a string.

Syntax
str.count(value, start, end), where :
value is the substring which is to be searched in the string [Required]
start is the starting index within the string where the search for the specified value starts [Optional]
end is the ending index within the string where the search for the specified value ends [Optional]
```

**Example 1**

```python
email = input("Enter Email: ")

if email.count("@") == 1:
    print("Valid Email Format")
else:
    print("Invalid Email Format")
```

**Example 2**

```python
sentence = 'She sells seashells by the seashore. The shells she sells are surely seashells'
sentence.count('seashells')
#sentence.count('seashells',9,25)
```

**Example 3**

```python
orders = "Laptop,Mouse,Laptop,Keyboard,Laptop"

count = orders.count("Laptop")

print("Laptop Orders:", count)
```

**Example 4**

```python
sentence = 'She sells seashells by the seashore. The shells she sells are surely seashells'
print("Total seashells:", sentence.count('seashells'))
```
