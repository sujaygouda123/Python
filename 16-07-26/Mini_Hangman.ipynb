```python
import random

words = ["python", "machine", "learning", "bridge"]

secret_word = random.choice(words)

guessed_word = ["_"] * len(secret_word)

guessed_letters = []
wrong_attempts = 0
max_attempts = 6

print("=" * 40)
print("           WORD GUESSING GAME")
print("=" * 40)
print(f"You have {max_attempts} wrong attempts.")
print()

while wrong_attempts < max_attempts and "_" in guessed_word:

    print("Word :", " ".join(guessed_word))
    print("Guessed Letters :", ", ".join(guessed_letters))
    print(f"Remaining Attempts : {max_attempts - wrong_attempts}")

    try:
        guess = input("Enter a letter: ").lower()

        if len(guess) != 1:
            print("Please enter only ONE letter.\n")
            continue

        if not guess.isalpha():
            print("Only alphabets are allowed.\n")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.\n")
            continue

        guessed_letters.append(guess)

        if guess in secret_word:
            print("Correct Guess!\n")

            for i in range(len(secret_word)):
                if secret_word[i] == guess:
                    guessed_word[i] = guess

        else:
            wrong_attempts += 1
            print("Wrong Guess!")
            print(f"Attempts Left: {max_attempts - wrong_attempts}\n")

    except Exception as e:
        print("Invalid Input:", e)

print("\n" + "=" * 40)

if "_" not in guessed_word:
    print("Congratulations! You guessed the word.")
    print("Word:", secret_word)
else:
    print("Game Over!")
    print("The correct word was:", secret_word)

print("=" * 40)
```
