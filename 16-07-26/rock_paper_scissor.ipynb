```python
from random import randint

def main():
    rules()
    
    user_name = name()
    
    words = ["rock", "paper", "scissors"]
    rematch = 'yes'
    
    while rematch == 'yes':
        bot, user = game(3, words, user_name)
    
        end_score(bot, user, user_name)
        
        rematch = end_game(user_name)
        
    print("^^^^ Thanks for playing ^^^^")


def game(n, words, name):
    
    bot_count = 0
    user_count = 0
    thisdict = {
      "paper": "rock",
      "rock": "scissors",
      "scissors": "paper"
    }
    
    for i in range(n):
        print(f"\n--- Round {i + 1} ---")
        
        bot = words[randint(0, 2)]
        user = user_input(words)
        print(f"Bot: {bot}")
        
        if user == bot:
            print("--- It's a Draw ---")
        elif thisdict[user] == bot:
            user_count += 1
            print(f"--- {name} Wins ---")
        else:
            bot_count += 1
            print("--- Bot Wins ---")
            
    return bot_count, user_count

def user_input(word):
    
    user = ""
    while user not in word:
        user = input("Enter rock, paper, or scissors: ").strip().lower()
        if user not in word:
            print("Invalid entry! Please enter: rock, paper, or scissors")
    return user


def end_score(bot_count, user_count, name):
   
    if user_count == bot_count:
        print('******   SUCK   ******')
    elif user_count > bot_count:
        print('******   HERO   ******')
    else:
        print('******   LOSER   ******')
    print(f"|{name:^10}| Bot |\n|{user_count:^10}|{bot_count:^10}|\n{'*' * 23}")


def end_game(name):
    
    user = ""
    while user not in ["yes", "no"]:
        user = input(f"Rematch, {name} ? yes or no ").strip().lower()
    return user
        

def name():
    
    try:
        name = input("What's your name? ").strip().title().split()
        print(f"Hello {name[0]}, let's play")
        return name[0]
        
    except IndexError:
        print("Hello Silent Ninja, let's play")
        return "Silent Ninja"


def rules():
    print("""| ROCK | PAPER | SCISSOR |
    
Rules:
1. Input your choice: 'rock', 'paper', or 'scissors'.
2. The Bot responds with a random counter-move.
3. Play repeats automatically for a 3-round loop.
4. The final game results are calculated and displayed.""")


main()
```
