### 1. While Loop
With the **while** loop we can execute a set of statements as long as a condition is `True`.

```python
# Example 1
correct_pin = "1234"
attempts = 3

while attempts > 0:
    pin = input("Enter ATM PIN: ")

    if pin == correct_pin:
        print("Login Successful")
        break

    attempts -= 1
    print("Incorrect PIN")
    print("Attempts Left:", attempts)

if attempts == 0:
    print("ATM Card Blocked")
```

```python
# Example 2
employee_count = 1

while employee_count <= 5:
    name = input(f"Enter Employee {employee_count} Name: ")
    print(name, "Attendance Recorded")
    employee_count += 1
```

```python
# Example 3
stock = 100

while stock > 0:
    sold = int(input("Enter Quantity Sold: "))
    stock -= sold
    print("Remaining Stock:", stock)

print("Out of Stock")
```

```python
# Example 4
username = ""

while username != "admin":
    username = input("Enter Username: ")

print("Access Granted")
```

### 2. For Loop
A **for** loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

```python
# Example 1
salaries = [35000, 42000, 28000, 55000, 61000]

total = 0

for salary in salaries:
    total += salary

print("Total Salary Paid:", total)
```

```python
# Example 2
marks = [78, 85, 64, 91, 55]

for mark in marks:
    print("Student Marks:", mark)
```

```python
# Example 3
prices = [250, 450, 180, 900]

bill = 0

for price in prices:
    bill += price

print("Final Bill =", bill)
```

```python
# Example 4
employees = ["john", "alice", "rohan"]

for employee in employees:
    print(employee + "@company.com")
```

### 3. If Else
The **if** block runs when the condition is `True`; otherwise the **else** block runs. Used for simple two-way decisions.

```python
# Example 1
salary = int(input("Monthly Salary: "))

if salary >= 50000:
    print("Loan Approved")
else:
    print("Loan Rejected")
```

```python
# Example 2
cart = float(input("Cart Amount: "))

if cart >= 999:
    print("Free Delivery")
else:
    print("Delivery Charge ₹99")
```

```python
# Example 3
experience = int(input("Years of Experience: "))

if experience >= 5:
    print("Bonus Eligible")
else:
    print("No Bonus")
```

```python
# Example 4
age = int(input("Age: "))

if age >= 18:
    print("Eligible to Apply")
else:
    print("Minor Applicant")
```

### 4. If Elif Else
**elif** lets you check multiple conditions in order — Python runs the first block whose condition is `True`, and falls back to **else** if none match.

```python
# Example 1
income = float(input("Annual Income: "))

if income <= 300000:
    print("No Tax")
elif income <= 700000:
    print("5% Tax")
elif income <= 1000000:
    print("10% Tax")
else:
    print("20% Tax")
```

```python
# Example 2
marks = int(input("Enter Marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
# Example 3
weight = float(input("Package Weight (kg): "))

if weight <= 1:
    print("Shipping Cost: ₹40")
elif weight <= 5:
    print("Shipping Cost: ₹90")
elif weight <= 10:
    print("Shipping Cost: ₹150")
else:
    print("Shipping Cost: ₹250")
```

```python
# Example 4
amount = float(input("Purchase Amount: "))

if amount >= 5000:
    print("20% Discount")
elif amount >= 2000:
    print("10% Discount")
elif amount >= 1000:
    print("5% Discount")
else:
    print("No Discount")
```

### 5. User Defined Function
A function, created with `def`, is a reusable block of code that only runs when it is called, and can accept parameters as input.

```python
# Example 1
def calculate_emi(principal, months):
    emi = principal / months
    print("Monthly EMI =", emi)

calculate_emi(500000, 60)
```

```python
# Example 2
def employee_id(name):
    print("EMP-" + name[:3].upper())

employee_id("Sujay")
```

```python
# Example 3
def simple_interest(principal, rate, years):
    interest = (principal * rate * years) / 100
    print("Interest =", interest)

simple_interest(50000, 8, 3)
```

```python
# Example 4
def calculate_area(length, breadth):
    area = length * breadth
    print("Area =", area)

calculate_area(12, 8)
```

### 6. Function + For Loop
A function can contain a **for** loop inside it, so a repeated task can be reused simply by calling the function.

```python
# Example 1
def employee_list(names):
    for employee in names:
        print(employee)

employees = ["John", "Alice", "Rahul", "David"]

employee_list(employees)
```

```python
# Example 2
def calculate_bill(prices):
    total = 0

    for price in prices:
        total += price

    gst = total * 0.18

    print("Total:", total)
    print("GST:", gst)
    print("Final:", total + gst)

calculate_bill([250, 800, 450])
```

```python
# Example 3
def total_marks(marks_list):
    total = 0
    for mark in marks_list:
        total += mark
    print("Total Marks:", total)

total_marks([78, 85, 64, 91])
```

```python
# Example 4
def print_invoice(items):
    for item in items:
        print("Item:", item)

print_invoice(["Pen", "Notebook", "Eraser"])
```

### 7. Function + For Loop + If
A function that loops through a sequence and applies an **if** condition to each item as it goes.

```python
# Example 1
def calculate_bonus(salaries):

    for salary in salaries:

        if salary >= 50000:
            print(salary, "Eligible")
        else:
            print(salary, "Not Eligible")

calculate_bonus([45000, 52000, 68000, 30000])
```

```python
# Example 2
def failed_students(marks):

    for mark in marks:

        if mark < 35:
            print(mark)

failed_students([80, 25, 91, 18, 45])
```

```python
# Example 3
def check_results(marks):
    for mark in marks:
        if mark >= 35:
            print(mark, "Pass")
        else:
            print(mark, "Fail")

check_results([20, 45, 60, 30])
```

```python
# Example 4
def stock_alert(stock_levels):
    for item in stock_levels:
        if stock_levels[item] < 10:
            print(item, "Low Stock")

stock_alert({"Pen": 5, "Notebook": 20, "Eraser": 8})
```

### 8. Break
The **break** statement immediately exits the loop, skipping any remaining iterations.

```python
# Example 1
orders = ["Laptop", "Mouse", "Cancelled", "Keyboard"]

for order in orders:

    if order == "Cancelled":
        break

    print(order)
```

```python
# Example 2
products = ["Phone", "TV", "AC", "Laptop"]

search = input("Search Product: ")

for product in products:

    if product == search:
        print("Product Found")
        break
    print("Product Not found")
```

```python
# Example 3
numbers = [10, 25, 30, -5, 40]

for num in numbers:
    if num < 0:
        break
    print(num)
```

```python
# Example 4
logins = ["Success", "Success", "Failed", "Success"]

for status in logins:
    if status == "Failed":
        print("Account Locked")
        break
    print("Login OK")
```

### 9. Continue
The **continue** statement skips the rest of the current iteration and moves on to the next one, without exiting the loop.

```python
# Example 1
products = {
    "Laptop": 5,
    "Mouse": 0,
    "Keyboard": 8,
    "Monitor": 0
}

for item in products:

    if products[item] == 0:
        continue

    print(item, "Available")
```

```python
# Example 2
attendance = ["Present", "Absent", "Present", "Absent", "Present"]

for status in attendance:

    if status == "Absent":
        continue

    print(status)
```

```python
# Example 3
numbers = [10, -5, 20, -8, 15]

for num in numbers:
    if num < 0:
        continue
    print("Valid Number:", num)
```

```python
# Example 4
users = ["john", "blocked_user", "alice", "blocked_user2"]

for user in users:
    if "blocked" in user:
        continue
    print("Notification Sent to", user)
```

```python
# Example 4
cities = {"Mumbai", "Delhi", "Bangalore"}

for city in cities:
    print(city)
```

### 11. For Loop with If
Combining a **for** loop with an **if** statement lets you filter a sequence, acting only on the items that meet a condition.

```python
# Example 1
employees = {
    "John":45000,
    "Alice":80000,
    "David":35000,
    "Rohan":95000
}

for name in employees:

    if employees[name] >= 50000:
        print(name, employees[name])
```

```python
# Example 2
numbers = [12, 7, 18, 25, 30, 3]

for num in numbers:
    if num % 2 == 0:
        print(num, "is Even")
```

```python
# Example 3
marks = [55, 30, 80, 25, 60]

for mark in marks:
    if mark >= 40:
        print(mark, "Pass")
```

```python
# Example 4
stock = {"Pen": 0, "Notebook": 12, "Eraser": 0, "Pencil": 5}

for item in stock:
    if stock[item] > 0:
        print(item, "In Stock")
```

### 12. For Loop with not in
**not in** is the opposite of `in` — it lets a loop skip or exclude items that belong to another collection (e.g. a blocklist).

```python
# Example 1
blocked = ["spam123", "fake001"]

users = ["john", "alice", "spam123", "rohan"]

for user in users:

    if user not in blocked:
        print(user)
```

```python
# Example 2
discontinued = ["Pager", "CD Player"]

products = ["Laptop", "Pager", "Smartwatch", "CD Player", "Tablet"]

for product in products:
    if product not in discontinued:
        print(product)
```

```python
# Example 3
failed_subjects = ["Physics", "Chemistry"]

all_subjects = ["Physics", "Maths", "Chemistry", "English"]

for subject in all_subjects:
    if subject not in failed_subjects:
        print(subject, "Passed")
```

```python
# Example 4
paid_customers = ["Ravi", "Meena"]

customers = ["Ravi", "Suresh", "Meena", "Anita"]

for customer in customers:
    if customer not in paid_customers:
        print("Send Payment Reminder to", customer)
```

### 13. If with in
Using **in** inside an `if` checks whether a value exists inside a string, list, or other collection — useful for substring/keyword checks.

```python
# Example 1
email = input("Enter Email: ")

if "@gmail.com" in email:
    print("Gmail Account")
else:
    print("Other Email")
```

```python
# Example 2
filename = input("Enter File Name: ")

if ".pdf" in filename.lower():
    print("PDF File")
```

```python
# Example 3
feedback = input("Enter Feedback: ")

if "good" in feedback.lower():
    print("Positive Feedback")
```

```python
# Example 4
letter = input("Enter a Letter: ")

if letter.lower() in "aeiou":
    print("Vowel")
else:
    print("Consonant")
```

### 14. If with not in
**not in** inside an `if` checks that a value is *absent* from a string or collection — useful when something must NOT be present.

```python
# Example 1
password = input("Enter Password: ")

if "@" not in password:
    print("Password must contain @")
else:
    print("Strong Password")
```

```python
# Example 2
username = input("Enter Username: ")

banned_words = ["admin", "root", "test"]

if username.lower() not in banned_words:
    print("Username Available")
else:
    print("Username Not Allowed")
```

```python
# Example 3
expired_codes = ["SAVE10", "OLD50"]

code = input("Enter Coupon Code: ")

if code not in expired_codes:
    print("Coupon Valid")
else:
    print("Coupon Expired")
```

```python
# Example 4
extension = input("Enter File Extension (e.g. .pdf): ")

disallowed = [".exe", ".bat", ".sh"]

if extension not in disallowed:
    print("File Allowed")
else:
    print("File Blocked")
```
